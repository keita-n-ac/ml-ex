2. 長方形の縦の長さと横の長さを読み込んで，その長方形の面積と周の長さを求めるプログラムを作成せよ

```python
a = int(input('縦の長さを入力してください: '))
b = int(input('縦の長さを入力してください: '))
print('面積: ', a * b)
print('周の長さ: ', 2 * a + 2 * b)
```
