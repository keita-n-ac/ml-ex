```python
for i in range(10):
    for j in range(10):
        if i == j:
            print(i, end='')
        else:
            print(' ', end='')
    print()
```

```python
n = int(input('段数を入力してください'))
for i in range(n):
    for j in range(n - i):
        print('*', end='')
    print()
```

```python
for i in range(10):
    for j in range(10):
        print((i + j) % 10, end='')
    print()
```

```python
n = int(input('段数を入力してください'))
for i in range(n):
    for j in range(n - i - 1):
        print(' ', end='')
    for k in range(2 * i + 1):
        print('*', end='')
    print()
```
