### 演習問題
1. 乱数を利用して，プログラムで10000回サイコロを振ったときに，各目がそれぞれ何回出たのかを出力するプログラムを作成せよ

```python
import random

counts = [0, 0, 0, 0, 0, 0]
# counts[0]に1が出た回数，counts[1]に2が出た回数，…，counts[5]に6が出た回数を保存する

for i in range(10000):
    r = random.randint(1, 6)
    counts[r - 1] += 1

for i in range(len(counts)):
    print(i + 1, 'が出た回数:', counts[i])
```

2. じゃんけんにおけるグーを0，チョキを1，パーを2として扱うと，乱数を使用してプログラムでじゃんけんを行うことができる，この説明を利用して簡単なじゃんけんゲームを各自で考え，プログラムで実装せよ

```python
import random

hand = int(input('0, 1, 2のどれかを入力してください（グー: 0，チョキ: 1，パー: 2）'))
enemy = random.randint(0, 2)
hands = {0: 'グー', 1: 'チョキ', 2: 'パー'}
print('あなたの手:', hands[hand], 'VS 相手の手:', hands[enemy])

# グーのとき
if hand == 0:
    if enemy == 0:
        print('あいこです')
    elif enemy == 1:
        print('あなたの勝ちです')
    else:
        print('あなたの負けです')
# チョキのとき
elif hand == 1:
    if enemy == 0:
        print('あなたの負けです')
    elif enemy == 1:
        print('あいこです')
    else:
        print('あなたの勝ちです')
# パーのとき
else:
    if enemy == 0:
        print('あなたの勝ちです')
    elif enemy == 1:
        print('あなたの負けです')
    else:
        print('あいこです')
        
```

3. サンプルプログラムを参考にして，比例のグラフ，反比例のグラフ，1次関数のグラフ，2次関数のグラフを作成せよ

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(0.01, 10, 0.01) # x軸の値
y1 = x          # y軸の値 (比例: y = x)
y2 = 1.0 / x    # y軸の値 (反比例: y = 1/x)
y3 = x + 1      # y軸の値 (反比例: y = x+1)
y4 = x * x      # y軸の値 (反比例: y = x^2)

plt.xlabel('x') # x軸の名前の設定
plt.ylabel('y') # y軸の名前の設定
plt.plot(x, y1) # 折れ線グラフの作成
plt.plot(x, y2) # 折れ線グラフの作成
plt.plot(x, y3) # 折れ線グラフの作成
plt.plot(x, y4) # 折れ線グラフの作成
plt.show()      # 作成したグラフの表⽰
```
