1. 以下の表に対して，横軸を区名，縦軸を人口数とした積み立て棒グラフを作成せよ

| Area       | Male   | Female | 
|------------|--------|--------| 
| Kita-ku    | 136596 | 152727 | 
| Minami-ku  | 62347  | 73430  | 
| Nishi-ku   | 100027 | 117013 | 
| Higashi-ku | 126023 | 139356 | 

```python
import matplotlib.pyplot as plt

num = [1, 2, 3, 4]
male = [136596, 62347, 100027, 126023]
female = [152727, 73430, 117013, 139356]
area = ['Kita-ku', 'Minami-ku', 'Nishi-ku', 'Higashi-ku']

plt.bar(num, male, tick_label=area, color='b')
plt.bar(num, female, bottom=male, color='r')
plt.legend(['Male', 'FeMale'])
plt.show()
```

2. 以下の指示に従って，プログラムを作成せよ
Step1: サンプルプログラムを利用して，-πからπの範囲を要素数が100となるように分割したデータdataを作成する（π: 円周率）
Step2: dataの要素値を余弦(cos)値に変換したものをデータxとする
Step3: dataの要素値を正弦(sin)値に変換したものをデータyとする
Step4: データxとデータyを散布図でプロットを行うと円状にプロットされる

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.linspace(-np.pi, np.pi, 100)
x = np.sin(data)
y = np.cos(data)
plt.scatter(x, y, marker='*', color='b')
plt.show()
```

3. データdata [10, 20, 30, 40]，ラベルlabel ['A', 'B', 'C', 'D']に対する円グラフを作成しなさい．ただし，円グラフで表示するラベルには円グラフで占める割合を書くこと．
  - 今回の場合，A を A : 10.0%，B を B : 20.0%，C を C : 30.0%，D を D : 40.0%と表示する．

```python
import matplotlib.pyplot as plt
data = [10, 20, 30, 40]
label = ['A', 'B', 'C', 'D']
total = sum(data)

for i in range(len(label)):
    label[i] = label[i] + ' : ' + str(data[i] / total * 100) + '%'

plt.pie(data, labels=label)
plt.axis("equal")
plt.show()
```
