### matplotlib応用（以降の説明ではmatplotlib.pyplot を pltとして扱っていることに注意）
- グラフの表示範囲を設定する（指定しない場合はデータに応じてプログラム側が表示範囲を自動的に決める）
  - plt.xlim([a, b]) : グラフを表示するx軸（幅）をaからbまでとする
  - plt.ylim([a, b]) : グラフを表示するy軸（高さ）をaからbまでとする

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 50, 50)
y = 0.5 * x + 1
plt.xlim([0, 10])
plt.ylim([0, 6])
plt.plot(x, y)
plt.show()
```

- グラフのタイトル・ラベルを指定する
  - plt.title('文字列') : グラフのタイトルを指定した文字列にする
  - plt.xlable('文字列') : x軸のラベルを指定した文字列にする
  - plt.ylable('文字列') : y軸のラベルを指定した文字列にする

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.1, 50, 100)
y = 1 / x
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.title('y = 1 / x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.show()
```

- グラフにグリッドをつける

### ラベル（凡例）の設定
- plt.plot(x, y, color='色指定', label='ラベル名') とすることで，グラフの色とラベルを指定できる
  - colorの設定が無くてもラベルを設定できる plt.plot(x, y, label='ラベル名')

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.1, 40, 100)
y1 = 1 / x
y2 = 0.05 * x
plt.xlim([0, 40])
plt.ylim([0, 2])
plt.title('Sample')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1, color='r', label='y = 1/x') # 赤色+ラベル設定プロット
plt.plot(x, y2, color='b', label='y = 0.05x') # 青色+ラベル設定プロット
plt.legend() # ラベルを表示
plt.show()
```

### numpy応用
- np.pi: 円周率（3.141592…）を利用できる
- np.sin(x): 各要素に対して正弦(sin)値を求める
- np.cos(x): 各要素に対して余弦(cos)値を求める
- np.tan(x): 各要素に対して正接(tan)値を求める
  - xはnumpyで格納している変数としている 

```python
import numpy as np
print(np.pi)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1, color='r', label='y = sin(x)') # 赤色+ラベル設定プロット
plt.plot(x, y2, color='b', label='y = cos(x)') # 青色+ラベル設定プロット
plt.legend() # ラベルを表示
plt.show()
```

### マーカーの種類の変更
- plt.plot(x, y, marker='指定子') とすることで，マーカーの設定をすることができる
  - 'o' : 円

  - 'p' : 五角形

- グラフに色をつける
  - plt.plot(x, y, marker='指定子', markerfacecolor='色指定') とすることで，マーカーの色を指定することができる

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
# 赤線，マーカー円，マーカーマゼンタ，ラベル設定プロット
plt.plot(x, y1, color='r', label='y = sin(x)', marker='o', markerfacecolor='m')
# 青線，マーカー四角，マーカーシアン，ラベル設定プロット
plt.plot(x, y2, color='b', label='y = cos(x)', marker='s', markerfacecolor='c')
plt.legend() # ラベルを表示
plt.show()
```

- 線の種類を変更する
  - plt.plot(x, y, linestyle='指定子', color='色指定', linewidth=数値) とすることで，線の種類，色，太さを変更することができる
    - linestyleの設定
      - '-'  : 実線
      - '--' : 破線
      - '-.' : 点入り破線
      - ':'  : 点線
    - colorの設定
      - 'r' : 赤
      - 'g' : 緑
      - 'b' : 青
      - 'c' : シアン（水色）
      - 'm' : マゼンタ（赤紫）

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
# 赤破線，線の太さ3，ラベル設定プロット
plt.plot(x, y1, color='r', linestyle='--', linewidth=3, label='y = sin(x)')
# 青点線，線の太さ8，ラベル設定プロット
plt.plot(x, y2, color='b', linestyle=':', linewidth=8, label='y = cos(x)')
plt.legend() # ラベルを表示
plt.show()
```

### 棒グラフの作成
- plt.bar(変数1, 変数2)とすることで，変数1と変数2におけるインデックスに対応した棒グラフを作成できる
  - この場合の変数は，リストやnumpyを使用する
  - 変数1が横軸，変数2が縦軸となる

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1, 10, 10)
y = 10 * x
plt.bar(x, y) # 棒グラフを作成
plt.show()    # 作成したグラフの表示
```

- 棒グラフにラベルをつけた場合，ラベルの名前を保存したリストなどを用意する
  - 変数1と変数2におけるインデックスに対応するようにラベル名前保存リストを用意すること 
- plt.bar(変数1, 変数2, tick_label=ラベル保存変数) とすることで，ラベルを追加することができる
- 棒の色も折れ線グラフと同様に設定できる
  - plt.bar(変数1, 変数2, color='色指定')

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                       # x軸
y = [100, 50, 20, 200]                 # y軸
l = ['AA', 'A', 'B', 'C']              # ラベル
plt.bar(x, y, tick_label=l, color='k') # 黒+ラベル付き棒グラフを作成
plt.show()                             # 作成したグラフの表示
```

### 積み上げ棒グラフの作成
- plt.bar(変数1, 変数2)の後に，plt.bar(変数1, 変数3, bottom=変数2)とすることで，変数1を横軸，変数2の上に変数3を積み上げた棒グラフとなる
  - bottom=変数2は『変数2の上に積み上げる』という意味
- plt.legend(ラベル名前格納変数) とすることで，凡例を付けることができる
  - プロットした順番がインデックスとなる

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                         # x軸
y1 = [10, 20, 30, 40]                    # y軸データ1
y2 = [50, 60, 70, 80]                    # y軸データ2
l = ['A', 'B', 'C', 'D']                 # ラベル
plt.bar(x, y1, tick_label=l, color='b')  # 青+ラベル付き棒グラフを作成
plt.bar(x, y2, bottom=y1, color='k')     # 黒棒グラフをy1の上に積み上げる
plt.legend(['Test A', 'Test B'])         # プロットした順番に凡例が付く
plt.show()                               # 作成したグラフの表示
```

```python
# 3つの積み上げ
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                           # x軸
y1 = [10, 20, 30, 40]                      # y軸データ1
y2 = [50, 60, 70, 80]                      # y軸データ2
y3 = [100, 100, 100, 100]                  # y軸データ3
l = ['A', 'B', 'C', 'D']                   # ラベル
plt.bar(x, y1, tick_label=l, color='b')    # 青+ラベル付き棒グラフを作成
plt.bar(x, y2, bottom=y1, color='k')       # 黒棒グラフをy1の上に積み上げる
plt.bar(x, y3, bottom=y2, color='g')       # 緑棒グラフをy2の上に積み上げる
plt.legend(['Test A', 'Test B', 'Test C']) # プロットした順番に凡例が付く
plt.show()                                 # 作成したグラフの表示
```

### ヒストグラムの作成
- plt.hist(データ) とすることでヒストグラムを作ることができる
  - データはリストやnumpyで格納されているものとする
- numpyは乱数を使ってデータを簡単に作成できる
  - numpyをnpとして使うとき
    - np.random.randn(n): 平均0，分散1（標準偏差1）のデータをn個作成し，numpyで格納する
    - np.random.randint(a, b, c): a以上b未満のデータをc個作成し，numpyで格納する

```python
import numpy as np
data1 = np.random.randn(10)
print(data1)
data2 = np.random.randint(1, 7, 70)
print(data2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(10000)
plt.hist(data)
plt.show()
```

- plt.hist(data, bins=整数値) とすることで，ヒストグラムのビン数（棒の数）を設定できる

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randint(1, 7, 10000)
plt.hist(data, bins=6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100000)
plt.hist(data, bins=200)
plt.show()
```

### 散布図の作成

  - 各データはリストやnumpyで格納されているものとする

```python
import numpy as np
import matplotlib.pyplot as plt
data1 = np.random.randn(100)
data2 = np.random.randn(100)
plt.scatter(data1, data2)
plt.show()
```

### 円グラフの作成
- plt.pie(データ) とすることで円グラフを作成できる
  - 各データはリストやnumpyで格納されているものとする
- plt.pie(データ)の後に，plt.axis("equal")と書かないと楕円になる
  - つまり，真円にしたい場合，plt.axis("equal")と書く必要がある

```python
import matplotlib.pyplot as plt
data = [50, 30, 10, 5, 5]
plt.pie(data)
plt.axis("equal")
plt.show()
```

### ラベルの付与
- plt.pie(データ, labels=ラベル) とすることでラベル付き円グラフを作成できる
  - ラベルはリストやnumpyで格納されているものとし，データのインデックスと合わせる必要がある


```python
import matplotlib.pyplot as plt
data = [50, 30, 10, 5, 5]
label = ['AA', 'A', 'B', 'C', 'X']
plt.pie(data, labels=label)
plt.axis("equal")
plt.show()
```
