### 演習1
- 適切なコレクションを使って，以下のプログラムを作成してください
  1. 1から1000の整数の中で，「2の倍数を保存するコレクション」と「5の倍数を保存するコレクション」を作成してください
  2. 1の結果から，1から1000の整数の中で，「10の倍数を保存するコレクション」を作成してください
  3. 1と2の結果から，1から1000の整数の中で，「2の倍数の整数の和」と「5の倍数の整数の和」と「10の倍数の整数の和」を求めてください

```python
# 1の解答例（リストの場合）
times2 = []
times5 = []
for i in range(1, 1001):
    if i % 2 == 0:
        times2.append(i)
    if i % 5 == 0:
        times5.append(i)
        
print('1から1000の整数の中で，2の倍数は')
print(times2)
print('1から1000の整数の中で，5の倍数は')
print(times5)

# 2の解答例（題意からセットを使用した方が簡単に書ける）
times2 = set(times2)
times5 = set(times5)
times10 = times2 & times5
print('1から1000の整数の中で，10の倍数は')
print(times10)

# 3の解答例（題意からリストを使用した方が簡単に書ける）
times2 = list(times2)
times5 = list(times5)
times10 = list(times10)
print('1から1000の整数の中で，2の倍数の整数の和は')
print(sum(times2))
print('1から1000の整数の中で，5の倍数の整数の和は')
print(sum(times5))
print('1から1000の整数の中で，10の倍数の整数の和は')
print(sum(times10))
```

### 演習2
- 9×9の表に掛け算の九九の結果を埋めるようなイメージを持つリストの入れ子構造を作成してくだい
- インデックス[n]でアクセスするとn+1の段のリストを獲得することができる
    - インデックス[0]でアクセスすると1の段のリストを獲得することができる
    - インデックス[1]でアクセスすると2の段のリストを獲得することができる
    - インデックス[2]でアクセスすると3の段のリストを獲得することができる
- インデックス[n][m]でアクセスすると(n+1)(m+1)の結果を獲得することができる
    - インデックス[0][0]でアクセスすると1*1の結果を獲得することができる
    - インデックス[1][1]でアクセスすると2*2の結果を獲得することができる
    - インデックス[2][2]でアクセスすると3*3の結果を獲得することができる

```python
kuku = [] # 初期化
for i in range(9):
    temp = [] # 入れ子用で使うリスト
    for j in range(9):
        temp.append((i + 1) * (j + 1))
    kuku.append(temp)

# 1行ずつ出力する
for i in kuku:
    print(i)
```
