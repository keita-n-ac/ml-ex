### ライブラリ
- Pythonにはライブラリとよばれる「プログラムの集まり」が用意されている
  - このライブラリを利用することで，様々な機能を持つプログラムを簡単に作成できるようになる
1. 標準ライブラリ
- Pythonをインストールした際に付属しているライブラリ
  - 例: ```math```（数学機能），```random```（乱数機能）
2. 外部ライブラリ
- 新たにダウンロードしてインストールが必要なライブラリ
  - 例: ```matplotlib```（グラフ機能），```pytorch```（深層学習）

### ライブラリの使用例
- ライブラリを導入する場合，最初に```import ライブラリ名```と書く
- ライブラリに含まれる機能を使用するには，```.```（ドット）を打ち，機能名を続けることで使用することができる

````python
# 累乗と平方根を求めるmathライブラリのpow()，sqrt()の使用例
import math # mathライブラリの導⼊
# math.pow(a, b) # aのb乗
print(math.pow(2, 5)) # 2の5乗
# math.sqrt(a) # ルートa
print(math.sqrt(2)) # 2の平⽅根（ルート2）
````

### ライブラリの使用例
- ```as``` を使用することで，ライブラリに別名をつけて使用できる

````python
# mathライブラリをmという別名をつけた場合における
# 累乗と平方根を求めるpow()，sqrt()の使用例
import math as m   # mathライブラリをmとして導⼊
print(m.pow(2, 5)) # 2の5乗
print(m.sqrt(2))   # 2の平⽅根（ルート2）
````

### randomライブラリ
- Pythonでランダムな処理が必要な場合に使用するランダムな数字のことを乱数とよぶ
  - ```random```ライブラリ: 乱数を扱う標準ライブラリ
- サイコロを振るなどの『確率』をプログラムで扱う場合，このような処理が必要となる
- ```random.randint(a, b)```
  - a ≦ n ≦ b であるようなランダムな整数nを作成する
- ```random.choice(リスト変数名)```
  - リストにある値をランダムに取り出すことができる

````python
import random
# ランダムに1以上6以下の整数を10回出⼒する
# サイコロを10回振るイメージ
# 結果は毎回異なる
for i in range(10):
    print(random.randint(1, 6))
````

````python
import random
lottery = ['1等', '2等', '3等', 'ハズレ']
# 実行するたびに結果は異なる
print(random.choice(lottery))
````

- 数値データのリストを```numpy```で扱う場合，**ベクトル**とよぶ
- 数値データのリストの入れ子を```numpy```で扱う場合，**行列**とよぶ

````python
import numpy as np
v_data = [1, 2, 3, 4, 5, 6, 7, 8]
data1 = np.array(v_data)
print('ベクトルの表示')
print(data1)

m_data = [[1, 2, 3, 4], [5, 6, 7, 8]]
data2 = np.array(m_data)
print('行列の表示')
print(data2)

t_data = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
data3 = np.array(t_data)
print('テンソルの表示')
print(data3)
````

- ```numpy```に格納されているデータは，簡単に演算ができるようになる

````python
import numpy as np
v_data = [1, 2, 3, 4, 5, 6, 7, 8]
data = np.array(v_data)

print(data)      # データを表示
print(data + 1)  # すべての要素に対して + 1
print(data - 3)  # すべての要素に対して - 3
print(data * 6)  # すべての要素に対して * 6
print(data / 10) # すべての要素に対して / 10
````

````python
import numpy as np
v_data1 = [1, 2, 3, 4]
v_data2 = [5, 6, 7, 8]

data1 = np.array(v_data1)
data2 = np.array(v_data2)

print(data1 + data2) # 0番目要素同士の足し算，1番目要素同士の足し算，…となる
print(data1 - data2) # 0番目要素同士の引き算，1番目要素同士の引き算，…となる
print(data1 * data2) # 0番目要素同士の掛け算，1番目要素同士の掛け算，…となる
print(data1 / data2) # 0番目要素同士の割り算，1番目要素同士の割り算，…となる
````

- ```np.arange()```を使うことで，```for```文で使用する```range()```の方式で，```numpy```の要素を初期化できる
  - ```np.arange(a)``` で ```0, 1, 2, …, a-1で初期化```
  - ```np.arange(a, b)``` で ```a, a+1, a+2, …, b-2, b-1 で初期化```
  - ```np.arange(a, b, c)```: ```c```の値は繰り返しを行う際の変数の更新量を意味する
    - **cが正の値**: aから開始して，bの値以上にならないように繰り返す
    - **cが負の値**: aから開始して，bの値以下にならないように繰り返す

````python
import numpy as np
data1 = np.arange(5)
print(data1)
data2 = np.arange(5, 11)
print(data2)
data3 = np.arange(5, 21, 3)
print(data3)
data4 = np.arange(10, -1, -1)
print(data4)
````

### スライス記法

### ラベル（凡例）の設定
- ```plt.plot(x, y, color='色指定', label='ラベル名')``` とすることで，グラフの色とラベルを指定できる
  - 色の設定が無くてもラベルを設定できる ```plt.plot(x, y, label='ラベル名')```
- ラベルを設定してプロットを行った後，```plt.legend()``` と行うことで，ラベルを表示できる

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.1, 40, 100)
y1 = 1 / x
y2 = 0.05 * x
plt.xlim([0, 40])
plt.ylim([0, 2])
plt.title('Sample')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1, color='r', label='y = 1/x') # 赤色+ラベル設定プロット
plt.plot(x, y2, color='b', label='y = 0.05x') # 青色+ラベル設定プロット
plt.legend() # ラベルを表示
plt.show()
````

### マーカーの種類の変更

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1, color='r', label='y = sin(x)', marker='o') # 赤色+マーカー円+ラベル設定プロット
plt.plot(x, y2, color='b', label='y = cos(x)', marker='s') # 青色+マーカー四角+ラベル設定プロット
plt.legend() # ラベルを表示
plt.show()
````

- マーカーに色をつける
  - ```plt.plot(x, y, marker='指定子', markerfacecolor='色指定')``` とすることで，マーカーの色を指定することができる
    - ```'r'``` : 赤
    - ```'g'``` : 緑
    - ```'b'``` : 青

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
# 赤線，マーカー円，マーカーマゼンタ，ラベル設定プロット
plt.plot(x, y1, color='r', label='y = sin(x)', marker='o', markerfacecolor='m')
# 青線，マーカー四角，マーカーシアン，ラベル設定プロット
plt.plot(x, y2, color='b', label='y = cos(x)', marker='s', markerfacecolor='c')
plt.legend() # ラベルを表示
plt.show()
````

- 線の種類を変更する
  - ```plt.plot(x, y, linestyle='指定子', color='色指定', linewidth=数値)``` とすることで，線の種類，色，太さを変更することができる
    - linestyleの設定
      - ```'-'```  : 実線
      - ```'--'``` : 破線
      - ```'-.'``` : 点入り破線
      - ```':'```  : 点線
    - colorの設定
      - ```'r'``` : 赤
      - ```'g'``` : 緑
      - ```'b'``` : 青
      - ```'c'``` : シアン（水色）
      - ```'m'``` : マゼンタ（赤紫）
      - ```'y'``` : 黄
      - ```'k'``` : 黒
      - ```'w'``` : 白
- プロットに関する設定は組み合わせが可能

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlim([-np.pi, np.pi])
plt.ylim([-1, 1])
plt.title('Trigonometric function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
# 赤破線，線の太さ3，ラベル設定プロット
plt.plot(x, y1, color='r', linestyle='--', linewidth=3, label='y = sin(x)')
# 青点線，線の太さ8，ラベル設定プロット
plt.plot(x, y2, color='b', linestyle=':', linewidth=8, label='y = cos(x)')
plt.legend() # ラベルを表示
plt.show()
````

### 棒グラフの作成
- ```plt.bar(変数1, 変数2)```とすることで，変数1と変数2におけるインデックスに対応した棒グラフを作成できる
  - この場合の変数は，リストやnumpyを使用する
  - 変数1が横軸，変数2が縦軸となる

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1, 10, 10)
y = 10 * x
plt.bar(x, y) # 棒グラフを作成
plt.show()    # 作成したグラフの表示
````

- 棒グラフにラベルをつけた場合，ラベルの名前を保存したリストなどを用意する
  - 変数1と変数2におけるインデックスに対応するようにラベル名前保存リストを用意すること
- ```plt.bar(変数1, 変数2, tick_label=ラベル保存変数)``` とすることで，ラベルを追加することができる
- 棒の色も折れ線グラフと同様に設定できる
  - ```plt.bar(変数1, 変数2, color='色指定')```

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                       # x軸
y = [100, 50, 20, 200]                 # y軸
l = ['AA', 'A', 'B', 'C']              # ラベル
plt.bar(x, y, tick_label=l, color='k') # 黒+ラベル付き棒グラフを作成
plt.show()                             # 作成したグラフの表示
````

### 積み上げ棒グラフの作成
- ```plt.bar(変数1, 変数2)```の後に，```plt.bar(変数1, 変数3, bottom=変数2)```とすることで，変数1を横軸，変数2の上に変数3を積み上げた棒グラフとなる
  - ```bottom=変数2```は『変数2の上に積み上げる』という意味
- ```plt.legend(ラベル名前格納変数)``` とすることで，凡例を付けることができる
  - プロットした順番がインデックスとなる

````python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                         # x軸
y1 = [10, 20, 30, 40]                    # y軸データ1
y2 = [50, 60, 70, 80]                    # y軸データ2
l = ['A', 'B', 'C', 'D']                 # ラベル
plt.bar(x, y1, tick_label=l, color='b')  # 青+ラベル付き棒グラフを作成
plt.bar(x, y2, bottom=y1, color='k')     # 黒棒グラフをy1の上に積み上げる
plt.legend(['Test A', 'Test B'])         # プロットした順番に凡例が付く
plt.show()                               # 作成したグラフの表示
````

````python
# 3つの積み上げ
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]                           # x軸
y1 = [10, 20, 30, 40]                      # y軸データ1
y2 = [50, 60, 70, 80]                      # y軸データ2
y3 = [100, 100, 100, 100]                  # y軸データ3
l = ['A', 'B', 'C', 'D']                   # ラベル
plt.bar(x, y1, tick_label=l, color='b')    # 青+ラベル付き棒グラフを作成
plt.bar(x, y2, bottom=y1, color='k')       # 黒棒グラフをy1の上に積み上げる
plt.bar(x, y3, bottom=y2, color='g')       # 緑棒グラフをy2の上に積み上げる
plt.legend(['Test A', 'Test B', 'Test C']) # プロットした順番に凡例が付く
plt.show()                                 # 作成したグラフの表示
````

### ヒストグラムの作成
- ```plt.hist(データ)``` とすることでヒストグラムを作ることができる
  - データはリストやnumpyで格納されているものとする
- numpyは乱数を使ってデータを簡単に作成できる
  - numpyをnpとして使うとき
    - ```np.random.randn(n)```: 平均0，分散1（標準偏差1）のデータをn個作成し，numpyで格納する
    - ```np.random.randint(a, b, c)```: a以上b未満のデータをc個作成し，numpyで格納する

````python
import numpy as np
data1 = np.random.randn(10)
print(data1)
data2 = np.random.randint(1, 7, 70)
print(data2)
````

````python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(10000)
plt.hist(data)
plt.show()
````

- ```plt.hist(data, bins=整数値)``` とすることで，ヒストグラムのビン数（棒の数）を設定できる

````python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randint(1, 7, 10000)
plt.hist(data, bins=6)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100000)
plt.hist(data, bins=200)
plt.show()
````

### 散布図の作成
- ```plt.scatter(データ1, データ2)``` とすることで散布図を作成できる
  - 各データはリストやnumpyで格納されているものとする
  - データ1が横軸，データ2がy軸となる

````python
import numpy as np
import matplotlib.pyplot as plt
data1 = np.random.randn(100)
data2 = np.random.randn(100)
plt.scatter(data1, data2)
plt.show()
````

- 散布図プロットの種類を変更する
  - ```plt.scatter(データ1, データ2, marker='指定子', color='色指定')``` とすることで，マーカーの種類，色を変更することができる
    - markerの設定
      - ```'o'``` : 円
      - ```'s'``` : 四角
      - ```'p'``` : 五角形
      - ```'*'``` : 星
      - ```'+'``` : プラス
      - ```'D'``` : ダイアモンド（ひし形）
    - colorの設定
      - ```'r'``` : 赤
      - ```'g'``` : 緑
      - ```'b'``` : 青
      - ```'c'``` : シアン（水色）
      - ```'m'``` : マゼンタ（赤紫）
      - ```'y'``` : 黄
      - ```'k'``` : 黒
      - ```'w'``` : 白
- プロットに関する設定は組み合わせが可能

````python
import numpy as np
import matplotlib.pyplot as plt
data1 = np.random.randn(100)
data2 = np.random.randn(100)
data3 = np.random.randn(100)
plt.scatter(data1, data2, marker='*', color='r')
plt.scatter(data1, data3, marker='D', color='b')
plt.show()
````

### 円グラフの作成
- ```plt.pie(データ)``` とすることで円グラフを作成できる
  - 各データはリストやnumpyで格納されているものとする
- ```plt.pie(データ)```の後に，```plt.axis("equal")```と書かないと楕円になる
  - つまり，真円にしたい場合，```plt.axis("equal")```と書く必要がある

````python
import matplotlib.pyplot as plt
data = [50, 30, 10, 5, 5]
plt.pie(data)
plt.axis("equal")
plt.show()
````

### ラベルの付与
- ```plt.pie(データ, labels=ラベル)``` とすることでラベル付き円グラフを作成できる
  - ラベルはリストやnumpyで格納されているものとし，データのインデックスと合わせる必要がある
### numpyとfor文
- numpy格納した要素もリストと同様にfor文でアクセスできる

### 文字列の結合
- ```+``` 演算子で行うことができる

````python
# numpy格納した要素もリストと同様にfor文でアクセスできる
import numpy as np
print('パターン1')
x = np.array([1, 2, 3, 4, 5, 6])
for i in x:
    print(i)

print('パターン2')
x = np.array([10, 20, 30, 40, 50, 60])
for i in range(len(x)):
    print(x[i])
````

````python
# 文字列の結合は + 演算子で行うことができる
moji = ['A', 'B', 'ABC', 'XYZ']
for i in range(len(moji)):
    moji[i] = str(i) + ' : ' + moji[i]

for i in moji:
    print(i)

print(moji)
````
