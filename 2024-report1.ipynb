# 2024年 機械学習演習レポート 1回目 (締め切り: 2024/11/14)
- 以下のプログラムを作成しなさい
- 各問を満たすプログラムをwordに貼り付けて，提出すること

1. 2つの整数を読み込んで，その和，差，積，商を出力するプログラムを作成しなさい．出力結果は以下を参考にすること．
- 9 と 3 を読み込んだ場合
```
9 と 3 の和は 12 です
9 と 3 の差は 6 です
9 と 3 の積は 27 です
9 と 3 の商は 3.0 です
```

2. 正の整数を入力し，接尾辞「-st」，「-nd」，「-rd」，「-th」をつけ，英語の序数として出力するプログラムを作成しなさい．条件は以下の通り．
- 2桁の整数 かつ 十の位が1の数は th をつける
- それ以外のときは，一の位が1のとき st ，2のとき nd ，3のとき rd ，それ以外のときは th をつける
  - 23を入力すると 23rd ，11を入力すると 11th と出力する

3. 1から100までの整数を1つ入力して，その数が5で割り切れる または 5が含まれている場合 ```Yes``` を出力し，そうで無い場合は，```No``` を出力すること．ただし，1から100ではない整数が入力された場合は, ```Error``` を出力すること．

4. 1から100まで順番に数を1行ずつ出力していく．ただし，その数が3で割り切れる または 3が含まれている場合，数字の代わりに ```Foon``` を出力するプログラムを作成しなさい．出力結果は以下を参考にすること．
