####  問題1
- ```データ [3, 5, 4, 4, 4, 6, 3, 4, 5, 6, 1, 5, 4, 4, 4, 1, 5, 5, 6, 3, 1, 2, 3, 6, 2, 3, 5, 2, 6, 5]```における，合計値，平均値，分散値，標準偏差値を求めなさい

````python
import math
x = [3, 5, 4, 4, 4, 6, 3, 4, 5, 6, 1, 5, 4, 4, 4, 1, 5, 5, 6, 3, 1, 2, 3, 6, 2, 3, 5, 2, 6, 5]

# 合計値を求める
print('合計値', sum(x))

# 平均値を求める
a = sum(x) / len(x)
print('平均値', a)

# 偏差値を求める
for i in range(len(x)):
    x[i] -= a

# 偏差値の2乗値を求める
for i in range(len(x)):
    x[i] *= x[i]

# 分散値を求める
v = sum(x) / len(x)
print('分散値', v)

print('標準偏差', math.sqrt(v))
````

#### 問題2
- コレクションを適切に使用して，1から10000までの整数の中で，『2でも3でも5でも7でも11でも割り切れない数』のコレクションを作成しなさい


````python
ans = []
for i in range(1, 10001):
    if i % 2 != 0 and i % 3 != 0 and i % 5 != 0 and i % 7 != 0 and i % 11 != 0:
        ans.append(i)
print(ans)
````

#### 問題3
- 題意を満たすヒストグラムを作成しなさい

````python
import numpy as np
import matplotlib.pyplot as plt

# 表が出た回数を保存するコレクションを用意する
counts = []

# 二重ループで『100回のコイン投げシミュレーションを1000回行う』
# シミュレーションを1000回行う
for n in range(1000):
    # 各コイン投げで表が出た回数を保存する変数を初期化する
    c = 0
    # 100回のコイン投げシミュレーション
    for x in range(100):
        # 1が出たら カウントを行う
        if np.random.randint(0, 2) == 1:
            c += 1
    # カウントした結果を，表が出た回数を保存するコレクションに追加する
    counts.append(c)

# 表が出た回数を保存するコレクションをヒストグラムで表示する
plt.hist(counts, bins=100)
plt.show()
````

#### 問題4
- 以下の説明に従って，題意を満たす散布図を作成しなさい

````python
import numpy as np
import matplotlib.pyplot as plt
num = 10000
data1 = []
data2 = []

# Step1
for i in range(num):
    data1.append(np.random.rand())
    data2.append(np.random.rand())

# Step2
x1 = []
x2 = []
y1 = []
y2 = []
for i in range(num):
    if data1[i] < data2[i]:
        x1.append(data1[i])
        y1.append(data2[i])
    else:
        x2.append(data1[i])
        y2.append(data2[i])

# 散布図の表示
plt.scatter(x1, y1, marker='*', color='r')
plt.scatter(x2, y2, marker='D', color='b')
plt.show()
````
