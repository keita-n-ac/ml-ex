### 演習問題
1. 乱数を利用して，プログラムで10000回サイコロを振ったときに，各目がそれぞれ何回出たのかを出力するプログラムを作成せよ

```python
import random

counts = [0, 0, 0, 0, 0, 0]
# counts[0]に1が出た回数，counts[1]に2が出た回数，…，counts[5]に6が出た回数を保存する

for i in range(10000):
    r = random.randint(1, 6)
    counts[r - 1] += 1

for i in range(len(counts)):
    print(i + 1, 'が出た回数:', counts[i])
```

2. サンプルプログラムを参考にして，比例のグラフ，反比例のグラフ，1次関数のグラフ，2次関数のグラフを作成せよ

```python


import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.01, 10, 0.01) # x軸の値
y1 = x          # y軸の値 (比例の例: y = x)
y2 = 1.0 / x    # y軸の値 (反比例の例: y = 1/x)
y3 = x + 1      # y軸の値 (1次関数の例: y = x+1)
y4 = x * x      # y軸の値 (2次関数の例: y = x^2)

plt.xlabel('x') # x軸の名前の設定
plt.ylabel('y') # y軸の名前の設定
plt.plot(x, y1, label='y = x') # 折れ線グラフの作成
plt.plot(x, y2, label='y = 1/x') # 折れ線グラフの作成
plt.plot(x, y3, label='y = x+1') # 折れ線グラフの作成
plt.plot(x, y4, label='y = x^2') # 折れ線グラフの作成
plt.legend()    # 凡例を表示
plt.show()      # 作成したグラフの表⽰

```

4. 以下の指示に従って，プログラムを作成せよ

- Step1: サンプルプログラムを利用して，-πからπの範囲を要素数が100となるように分割したデータdataを作成する（π: 円周率）
- Step2: dataの要素値を余弦(cos)値に変換したものをデータxとする
- Step3: dataの要素値を正弦(sin)値に変換したものをデータyとする
- Step4: データxとデータyを散布図でプロットを行うと円状にプロットされる

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.linspace(-np.pi, np.pi, 100)
x = np.sin(data)
y = np.cos(data)
plt.scatter(x, y, marker='*', color='b')
plt.show()
```

6. 以下の指示に従って，プログラムを作成せよ
- Step1: データ1（data1）を np.random.randn(1000)，データ2（data2）を np.random.randn(1000)で作成しなさい
- Step2: 横軸をデータ1の値，縦軸をデータ2の値として散布図を作成する（ただし，データ1の値，データ2の値が両方0以上の場合は，マーカーと色を変えて表示すること）
  - ヒント: データ1とデータ2を，両方0以上の場合と，そうでない場合にデータを分割する

```python
import numpy as np
import matplotlib.pyplot as plt
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

x1 = []
x2 = []
y1 = []
y2 = []
for i in range(1000):
    if data1[i] >= 0 and data2[i] >= 0:
        x1.append(data1[i])
        y1.append(data2[i])
    else:
        x2.append(data1[i])
        y2.append(data2[i])

plt.scatter(x1, y1, marker='*', color='r')
plt.scatter(x2, y2, marker='D', color='b')
plt.show()
```
