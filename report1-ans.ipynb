```python
num1 = int(input('1つ目の整数を入力してください'))
num2 = int(input('2つ目の整数を入力してください'))

print(num1, 'と', num2, 'の和は', num1 + num2, 'です')
print(num1, 'と', num2, 'の差は', num1 - num2, 'です')
print(num1, 'と', num2, 'の積は', num1 * num2, 'です')
print(num1, 'と', num2, 'の商は', num1 / num2, 'です')
```

```python
num = int(input('正の整数を入力してください'))

if num >= 10 and num // 10 == 1:
    print(num, 'th')
elif num % 10 == 1:
    print(num, 'st')
elif num % 10 == 2:
    print(num, 'nd')
elif num % 10 == 3:
    print(num, 'rd')
else:
    print(num, 'th')
```

```python
for i in range(1, 101):
    if i % 3 == 0 or i // 10 == 3 or i % 10 == 3:
        print('Foon')
    else:
        print(i)
```

```python
n = int(input('段数を入力してください'))
for i in range(n):
    for j in range(n - i - 1):
        print('  ', end='')
    for k in range(2 * i + 1):
        print('**', end='')
    print()
```
