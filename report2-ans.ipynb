1. データ [3, 5, 4, 4, 4, 6, 3, 4, 5, 6, 1, 5, 4, 4, 4, 1, 5, 5, 6, 3, 1, 2, 3, 6, 2, 3, 5, 2, 6, 5] における，合計値，平均値，分散値，標準偏差値を求めなさい

```python
import math
x = [3, 5, 4, 4, 4, 6, 3, 4, 5, 6, 1, 5, 4, 4, 4, 1, 5, 5, 6, 3, 1, 2, 3, 6, 2, 3, 5, 2, 6, 5]

# 合計値を求める
print('合計値', sum(x))

# 平均値を求める
a = sum(x) / len(x)
print('平均値', a)

# 偏差値を求める
for i in range(len(x)):
    x[i] -= a

# 偏差値の2乗値を求める
for i in range(len(x)):
    x[i] *= x[i]

# 分散値を求める
v = sum(x) / len(x)
print('分散値', v)

print('標準偏差', math.sqrt(v))
```

3. 以下の説明に従って，題意を満たすヒストグラムを作成しなさい
- 100回のコイン投げシミュレーションを1000回行い，横軸を『表が出た回数』，縦軸を『1000回中に起こった回数』となるようなヒストグラムを作成しなさい

```python
import numpy as np
import matplotlib.pyplot as plt
counts = []

# シミュレーションを1000回行う
for n in range(1000):
    c = 0
    # 100回のコイン投げシミュレーション
    for x in range(100):
        if np.random.randint(0, 2) == 1:
            c += 1
    counts.append(c)

# ヒストグラムの表示
plt.hist(counts, bins=100)
plt.show()
```

4. 以下の説明に従って，題意を満たすヒストグラムを作成しなさい
- Step1: サンプルプログラム4-2を参考にして，データ1（data1）に要素数10000個の乱数を追加することで作成しなさい，データ2（data2）も同様に作成しなさい
- Step2: 横軸をデータ1の値，縦軸をデータ2の値として散布図を作成する（ただし，データ2の値がデータ1の値よりも小さい場合とそうでない場合は区別できるように，マーカーと色を変えて表示すること）

```python
import numpy as np
import matplotlib.pyplot as plt
num = 10000
data1 = []
data2 = []

# Step1
for i in range(num):
    data1.append(np.random.rand())
    data2.append(np.random.rand())

# Step2
x1 = []
x2 = []
y1 = []
y2 = []
for i in range(num):
    if data1[i] < data2[i]:
        x1.append(data1[i])
        y1.append(data2[i])
    else:
        x2.append(data1[i])
        y2.append(data2[i])

# 散布図の表示        
plt.scatter(x1, y1, marker='*', color='r')
plt.scatter(x2, y2, marker='D', color='b')
plt.show()
```
