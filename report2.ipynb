#### 問題3 
  - 以下の説明に従って，題意を満たすヒストグラムを作成しなさい

```python
# サンプルプログラム
import numpy as np

for i in range(10):
    print(np.random.randint(1, 7))
```

```python
# サンプルプログラム4-1
import numpy as np

for i in range(10):
    print(np.random.rand())
```
